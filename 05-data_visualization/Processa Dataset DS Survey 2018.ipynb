### Dataset obtido do desafio do Kaggle: 2018 Kaggle ML & DS Survey Challenge

O Dataset foi resultado de de uma abrangente pesquisa conduzida para mapear o estado da área de data science e machine learning na indústria. Essa pesquisa foi conduzida em outubro de 2018, e teve como resultado 23.859 resposta

Os resultados incluem informações desde informações do que está acontecendo na área até as melhores maneiras para novos data scientists entrarem no mercado. Os dados foram publicados no formato mais bruto possível sem comprometer o anonimato dos entrevistados.

Link para o desafio: <https://www.kaggle.com/kaggle/kaggle-survey-2018>

### Leitura dos dados

```python
df2018 = pd.read_csv('data/DS_Survey2018/multipleChoiceResponses.csv')
print(df2018.shape)
df2018.head(2)
```

```python
new_names = [a + '_' + b for a, b in df2018.iloc[0, :].items()]
df2018.columns = new_names
df2018 = df2018.drop(0)
df2018.head(2)
```

```python
print("Total de usuários que responderam o questionario: {} - Total de colunas com perguntas: {}".format(*df2018.shape))
```

```python
questions = pd.read_csv('data/DS_Survey2018/SurveySchema.csv')
questions = questions.drop(columns=questions.columns[0])[:1].transpose()
questions = questions.reindex(['Q{}'.format(i) for i in range(1, 51)])
questions
```

```python
df = pd.concat([df2018.filter(regex='Q6'), df2018.filter(regex='Q22'),
               df2018.filter(regex='Q17'), df2018.filter(regex='Q34')], axis=1)
df.head()
```

```python
col_name = ['Q22_OTHER_TEXT_Of the choices that you selected in the previous question, which specific data visualization library or tool have you used the most? - Other - Text',
            'Q6_OTHER_TEXT_Select the title most similar to your current role (or most recent title if retired): - Other - Text',
            'Q17_OTHER_TEXT_What specific programming language do you use most often? - Other - Text',
            'Q34_OTHER_TEXT_During a typical data science project at work or school, approximately what proportion of your time is devoted to the following? (Answers must add up to 100%) - Other']
df.drop(col_name, axis=1, inplace=True)
df.head(2)
```

```python
time_ds_project = ['proportion_time-' + col.split(') - ')[-1] for col in df.columns[3:]]
info_user = ['current_role', 'data_vis_lib', 'prog_language']
df.columns = info_user + time_ds_project
# df = df.reset_index().rename(columns={'index': 'user_id'})
df[time_ds_project] = df[time_ds_project].astype(float)
df.head(2)
```

```python
df.to_csv(os.path.join("data", "ds_survey2018_processado.csv"), index=False)
```
